```python
def greet(): # function definition
    print("hello user")
    print("welcome to code world")
    print("have a nice day")
greet()#function calling
print("see you soon again")    
```

```python
def num(a):
    if a%2 == 0:
      print("even number")
    elif a%2 != 0:
       print("odd number")  
num(7)       
```

```python
#non-return
def sum(a,b):
    print(a+b)

result = sum(10,20)  
print(result)
```

```python
#return
def sum(a,b):
    return a+b

result = sum(10,20)  
print(result)
```

```python
#Default parameterised function
def country(country = 'INDIA'):
    print(f'I LOVE {country}')

country('UK') 
country('USA')
country()   
```
