```python
#accessing value using keys
#1st method
Student['Name']
Student['age']
```

```python
#2nd method
Student.get('Name')
Student.get('age')
```

```python
print(Student.keys())
print(Student.values())
print(Student.items())
```

```python
#adding anew pair
Student['city'] = 'lko'
print(Student)
```

```python
#deleting a pair
del Student['city']
print(Student)
```

```python
#update
Student['course'] = 'BBA'
print(Student)
```

```python
for i,v in Student.items():
  print (f'{i}:{v}')
```

```python
a = 10
b = 20
print("sum is",a+b)
print("sum of",a,"and",b,"is",a+b)
print(f"sum of {a} and {b} is {a+b}")
```

```python
#nested dictionary
company = {
    "emp001":{
        "name":"roy",
        "salary":40000,
        "department":"HR"
    },
    "emp002":{
        "name":"arpita",
        "salary":50000,
        "department":"IT"
    }
}
print(company)
```

```python
company["emp003"] = {
 'name':"shreya",
 "salary":45000,
 "department":"IT"
}
print(company)
```

```python
company["emp002"]["name"] = "raj"
print(company)
```
