```python
apps = ['whatsapp','chrome','google','galary','camera','spotify','pinterest','youtube','drive','rapido']
print(apps)
```

INDEXING AND SLICING

```python
print(apps[1])
print(apps[5])
print(apps[-2])
print(apps[-5])
```

```python
print(apps[1:6])
print(apps[1:6:2])
print(apps[:6])
print(apps[5:])
print(apps[:])
```

LIST METHODS.
- append - adds an element at the end of the list
- insert- adds an element at any specific position
- pop - delete an element from the end and also using index value
-  remove- delete an element using name of it
- copy- make a copy of that list
- sort- sort the elemsnts in specific order
- extend- combine two list

```python
apps.append('Hotstar')
print(apps)
```

```python
apps.insert(2,'Snapchat')
print(apps)
```

```python
apps.pop()#deletes last element
print(apps)
```

```python
apps.pop(0)
print(apps)
```

```python
apps.remove('drive')
print(apps)
```

```python
apps2 = apps.copy()
print(apps2)
print(apps)
```

```python
apps2.clear()
print(apps2)
```

```python
apps.sort()
print(apps)
```

```python
lst = [1,2,3,4,5]
apps.extend(lst)
print(apps)
```

```python
lst.extend(apps)
print(lst)
```
